## Selección de Características

```python
# Cargar los datos preproceasados de Orange valores no inputados son '-'
df = pd.read_csv('data/Saber_11__2019-2_preprocessed_v2.csv', sep=',', na_values='-')
df.head()
```

Hay 83 características en el conjunto de datos, incluyendo la variable objetivo. Para seleccionar las características más relevantes, se utilizarán dos criterios:
1. Los criterios que se utilizaron en los talleres uno y dos (Rank a partir de diferentes métricas).
2. Se utilizará un Random Forest para evaluar la importancia de las características.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

rank_orange = [
    "COLE_AREA_UBICACION", "COLE_BILINGUE", "COLE_CALENDARIO", "COLE_CARACTER",
    "COLE_CODIGO_ICFES", "COLE_DEPTO_UBICACION", "COLE_GENERO", "COLE_JORNADA",
    "COLE_NATURALEZA", "COLE_SEDE_PRINCIPAL", "DESEMP_INGLES", "ESTU_COD_RESIDE_MCPIO",
    "ESTU_DEDICACIONINTERNET", "ESTU_DEDICACIONLECTURADIARIA", "ESTU_DEPTO_PRESENTACION",
    "ESTU_DEPTO_RESIDE", "ESTU_ESTADOINVESTIGACION", "ESTU_ETNIA", "ESTU_GENERACION-E",
    "ESTU_GENERO", "ESTU_HORASSEMANATRABAJA", "ESTU_NACIONALIDAD", "ESTU_PAIS_RESIDE",
    "ESTU_PRIVADO_LIBERTAD", "ESTU_TIENEETNIA", "ESTU_TIPOREMUNERACION",
    "FAMI_COMECEREALFRUTOSLEGUMBRE", "FAMI_CUARTOSHOGAR", "FAMI_EDUCACIONMADRE",
    "FAMI_EDUCACIONPADRE", "FAMI_ESTRATOVIVIENDA", "FAMI_NUMLIBROS", "FAMI_PERSONASHOGAR",
    "FAMI_SITUACIONECONOMICA", "FAMI_TIENEAUTOMOVIL", "FAMI_TIENECOMPUTADOR",
    "FAMI_TIENECONSOLAVIDEOJUEGOS", "FAMI_TIENEHORNOMICROOGAS", "FAMI_TIENEINTERNET",
    "FAMI_TIENELAVADORA", "FAMI_TIENEMOTOCICLETA", "FAMI_TIENESERVICIOTV",
    "FAMI_TRABAJOLABORMADRE", "FAMI_TRABAJOLABORPADRE", "PERIODO"
]

df_copy_for_dict = df.copy()

# Label encoding
for col in rank_orange:
    if df[col].dtype == 'object':
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))

# Dividir los datos en características y objetivo
X = df[rank_orange]
y = df['PUNT_GLOBAL_CAT']

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo de Random Forest
forest = RandomForestClassifier(random_state=42)
forest.fit(X_train, y_train)

# Obtener la importancia de las características
importances = forest.feature_importances_
features_df = pd.DataFrame({'Feature': rank_orange, 'Importance': importances})
features_df = features_df.sort_values(by='Importance', ascending=False)

# Mostrar las características más importantes
features_df.head(20)
```

```python
# Lista de las 20 características más importantes 
important_features = [
    "DESEMP_INGLES", "COLE_CODIGO_ICFES", "COLE_JORNADA", "ESTU_GENERACION-E",
    "FAMI_NUMLIBROS", "FAMI_EDUCACIONMADRE", "ESTU_COD_RESIDE_MCPIO",
    "FAMI_EDUCACIONPADRE", "FAMI_TRABAJOLABORPADRE", "FAMI_ESTRATOVIVIENDA",
    "FAMI_TRABAJOLABORMADRE", "ESTU_DEPTO_PRESENTACION", "COLE_DEPTO_UBICACION",
    "ESTU_DEPTO_RESIDE", "FAMI_TIENECOMPUTADOR", "ESTU_HORASSEMANATRABAJA",
    "ESTU_DEDICACIONINTERNET", "FAMI_CUARTOSHOGAR", "COLE_NATURALEZA",
    "ESTU_DEDICACIONLECTURADIARIA","PUNT_GLOBAL_CAT"
]

# Crear un nuevo DataFrame con solo las características importantes
new_df = df[important_features]

df_copy_for_dict_no_encoded = df_copy_for_dict[important_features] 
df_copy_for_dict_encoded = df[important_features]

# Mostrar las primeras filas del nuevo DataFrame
new_df.head()
```

Debido a que se creo un nuevo dataframe usando las variables más importantes, ahora existen registros que son iguales, porque lo único que los diferenciaba eran las características que se eliminaron. Por lo tanto, se eliminan los registros duplicados.

```python
# Eliminar registros duplicados
new_df = new_df.drop_duplicates()
df_copy_for_dict_no_encoded = df_copy_for_dict_no_encoded.drop_duplicates()
df_copy_for_dict_encoded = df_copy_for_dict_encoded.drop_duplicates()
```

Con las características seleccionadas se construirán los modelos de clasificación supervisada. Véase que características relativas al puntaje obtenido en el ICFES no fueron escogidas, porque no tiene sentido utilizarlas para predecir el éxito estudiantil en el ICFES.

## Modelado

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import precision_score, recall_score, f1_score
import seaborn as sns
import xgboost as xgb

# Dividir el conjunto de datos
X_train, X_test, y_train, y_test = train_test_split(new_df.drop('PUNT_GLOBAL_CAT', axis=1), new_df['PUNT_GLOBAL_CAT'], test_size=0.3, random_state=42)

# Diccionario de modelos
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "LightGBM": LGBMClassifier(),
    # "XGBoost": XGBClassifier()
}

# Diccionario para almacenar las métricas
model_metrics = {
    'Model': [],
    'Precision': [],
    'Recall': [],
    'F1 Score': [],
}

# Entrenar y evaluar los modelos
for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Model: {name}")
    print(classification_report(y_test, y_pred, target_names=['Clase 1', 'Clase 2', 'Clase 3']))
    model_metrics['Model'].append(name)
    model_metrics['Precision'].append(precision_score(y_test, y_pred, average='weighted'))
    model_metrics['Recall'].append(recall_score(y_test, y_pred, average='weighted'))
    model_metrics['F1 Score'].append(f1_score(y_test, y_pred, average='weighted'))
    
    # # Matriz de confusión
    # cm = confusion_matrix(y_test, y_pred)
    # plt.figure(figsize=(6,6))
    # plt.title(f"Confusion Matrix for {name}")
    # sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', xticklabels=['Clase 1', 'Clase 2', 'Clase 3'], yticklabels=['Clase 1', 'Clase 2', 'Clase 3'])
    # plt.ylabel('Actual')
    # plt.xlabel('Predicted')
    # plt.show()

# XGBoost
# Convertir y a numérico (0, 1, 2)
le = LabelEncoder()
y = le.fit_transform(new_df['PUNT_GLOBAL_CAT'].values)

# Dividir el conjunto de datos
X_train, X_test, y_train, y_test = train_test_split(new_df.drop('PUNT_GLOBAL_CAT', axis=1), y, test_size=0.3, random_state=42)


# Entrenar el modelo
model = xgb.XGBClassifier(objective='multi:softmax', num_class=3)
model.fit(X_train, y_train)

# Predecir
y_pred = model.predict(X_test)

# Calcular métricas
model_metrics['Model'].append('XGBoost')
model_metrics['Precision'].append(precision_score(y_test, y_pred, average='weighted'))
model_metrics['Recall'].append(recall_score(y_test, y_pred, average='weighted'))
model_metrics['F1 Score'].append(f1_score(y_test, y_pred, average='weighted'))

# Convertir a DataFrame
metrics_df = pd.DataFrame(model_metrics)

# Gráfico de barras para comparar las métricas
plt.figure(figsize=(5, 5))
sns.barplot(x='F1 Score', y='Model', data=metrics_df, palette='mako')
plt.title('Comparación del F1 Score entre Modelos')
plt.show()

plt.figure(figsize=(5, 5))
sns.barplot(x='Precision', y='Model', data=metrics_df, palette='viridis')
plt.title('Comparación de Precision entre Modelos')
plt.show()

plt.figure(figsize=(5, 5))
sns.barplot(x='Recall', y='Model', data=metrics_df, palette='rocket')
plt.title('Comparación de Recall entre Modelos')
plt.show()
```

```python
# Cross Validation con el mejor modelo: XGBoost
from sklearn.model_selection import cross_val_score

prediction_mapping = {'bajo': 0, 'medio': 1, 'alto': 2}
new_df['PUNT_GLOBAL_NUM'] = new_df['PUNT_GLOBAL_CAT'].map(prediction_mapping)

# Dividir el conjunto de datos
X_train, X_test, y_train, y_test = train_test_split(new_df.drop(['PUNT_GLOBAL_CAT', 'PUNT_GLOBAL_NUM'], axis=1), new_df['PUNT_GLOBAL_NUM'], test_size=0.3, random_state=42)

# Entrenar el modelo
model = xgb.XGBClassifier(objective='multi:softmax', num_class=3)
model.fit(X_train, y_train)

# Cross Validation
scores = cross_val_score(model, X_train, y_train, cv=50, scoring='f1_weighted') # cv es el número de pliegues
print(f"Cross Validation Scores: {scores}")
print(f"Mean Cross Validation Score: {scores.mean()}")
print(f"Standard Deviation: {scores.std()}")
```

```python
from sklearn.metrics import roc_auc_score

# Predecir las probabilidades en lugar de las clases
y_pred_proba = model.predict_proba(X_test)

# Calcular el AUC utilizando 'ovr' (one-vs-rest) para el multi-class
auc = roc_auc_score(y_test, y_pred_proba, multi_class='ovr')
print(f"AUC: {auc}")
```

```python
from sklearn.metrics import roc_curve, auc

# Calcular y trazar las curvas ROC para cada par único de clases
n_classes = 3
fpr = dict()
tpr = dict()
roc_auc_dict = dict() 

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test == i, y_pred_proba[:, i])
    roc_auc_dict[i] = auc(fpr[i], tpr[i])

# Trazar las curvas ROC para cada clase
plt.figure(figsize=(8, 6))
colors = ['blue', 'red', 'green']
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i], lw=2, label=f'ROC curve (area = {roc_auc_dict[i]:.2f}) for Class {i}')
    print(f"ROC AUC Score for Class {i}: {roc_auc_dict[i]}")
    
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curvas ROC para Clases Múltiples')
plt.legend(loc='lower right')
plt.show()
```

```python
def csv_to_json_dict(csv_filename):
    # Leer el archivo CSV en un DataFrame de pandas
    df = pd.read_csv(csv_filename)
    
    # Convertir el DataFrame a una lista de registros (diccionarios)
    records = df.to_dict(orient='records')
    
    # Tomar el primer registro como un diccionario
    json_dict = records[0]

    # Asegurar que la clave ESTU_HORASSEMANATRABAJA sea un string
    json_dict['ESTU_HORASSEMANATRABAJA'] = str(json_dict['ESTU_HORASSEMANATRABAJA'])
    
    return json_dict

# Llamar a la función para convertir el archivo CSV a un diccionario
data_bajo = csv_to_json_dict("data/bajo.csv")
data_medio = csv_to_json_dict("data/medio.csv")
data_alto = csv_to_json_dict("data/alto.csv")
```
